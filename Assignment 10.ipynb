## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
# Using the between method
total_age_between = df[df['age'].between(17, 48)].shape[0]
print(f'Total number of persons having age between 17 to 48 (inclusive) using between method: {total_age_between}')
```

```python
# Without using the between method
total_age_between_without = df[(df['age'] >= 17) & (df['age'] <= 48)].shape[0]
print(f'Total number of persons having age between 17 to 48 (inclusive) without using between method: {total_age_between_without}')
```

## ***18. Which Workclass getting The Highest Salary***

```python
# Convert income to numeric values (if not already done)
# 0 for <=50K and 1 for >50K
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

# Calculate the average salary by workclass
highest_salary_workclass = df.groupby('workclass')['income'].mean().idxmax()
highest_salary_value = df.groupby('workclass')['income'].mean().max()

print(f'The workclass with the highest average salary is: {highest_salary_workclass} with an average salary of: {highest_salary_value:.2f}')
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
# Convert income to numeric values (if not already done)
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

# Calculate the proportion of each gender earning >50K
income_distribution = df.groupby('gender')['income'].mean()

# Calculate the chances
male_chance = income_distribution['Male']
female_chance = income_distribution['Female']

print(f"Chance of earning >50K for Males: {male_chance:.2f}")
print(f"Chance of earning >50K for Females: {female_chance:.2f}")

# Determine who has a better chance
if male_chance > female_chance:
    print("Males have a better chance of earning >50K.")
elif female_chance > male_chance:
    print("Females have a better chance of earning >50K.")
else:
    print("Both genders have the same chance of earning >50K.")
```
